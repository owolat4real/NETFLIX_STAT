```python
from scipy import stats

# Create a new column
neflix = neflix.copy()
neflix["post2015"] = neflix['release_year'] > 2015

# Split the data
before_2015 = neflix[neflix['post2015'] == False]['average_rating'].dropna()
after_2015 = neflix[neflix['post2015'] == True]['average_rating'].dropna()

# Show sample sizes
print(f"Samples: Before 2015 = {len(before_2015)}, After 2015 = {len(after_2015)}")

# T-test
t_test, p_value = stats.ttest_ind(before_2015, after_2015)
print(f't_test: {t_test}, P_value: {p_value}')

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in average ratings before and after 2015.")
else:
    print("No significant difference in average ratings before and after 2015.")
```

```python
from scipy import stats
import pandas as pd

# Ensure 'minute_of_life-changing_insight' is numeric
neflix['minute_of_life-changing_insight'] = pd.to_numeric(
    neflix['minute_of_life-changing_insight'], errors='coerce'
)

# Drop missing values in required columns
df = neflix.dropna(subset=['minute_of_life-changing_insight', 'insight_dec'])

# Group the data by 'insight_dec'
grouped_data = [
    group['minute_of_life-changing_insight'].values
    for name, group in df.groupby('insight_dec')
    if len(group) > 1  # make sure each group has more than 1 value
]

# One-way ANOVA (only if at least two valid groups exist)
if len(grouped_data) >= 2:
    f_stat, p_value = stats.f_oneway(*grouped_data)
    print("ANOVA Test: Effect of Insight Decision on Minute of Life-Changing Insight")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
else:
    print("Not enough valid groups for ANOVA.")
```
